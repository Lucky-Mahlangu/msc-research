## Imports

```python
import hashlib
import pandas as pd
```

```python
clean_instagram_data = pd.read_excel('../data/ground_truth/xlsx_data/instagram_sample_data.xlsx')
```

```python
clean_instagram_data
```

```python
df_anonymized = clean_instagram_data.copy()
```

```python
df_anonymized['anonymizedInstagramHandle'] = df_anonymized['instagramHandle'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest()[:7])
```

```python
#df_anonymized.to_excel('../data/ground_truth/xlsx_data/instagram_sample_data_partial_anonymized.xlsx', index = False)
```

```python
new_columns = ['anonymizedInstagramHandle', 'topK', 'domain',
       'topInfluencer', 'biography',
       'externalUrl', 'externalUrlShimmed', 'followersCount', 'followsCount',
       'hasChannel', 'highlightReelCount', 'isBusinessAccount',
       'joinedRecently', 'businessCategoryName', 'private', 'verified',
       'igtvVideoCount', 'relatedProfiles',
       'latestIgtvVideos', 'postsCount', 'latestPosts']
```

```python
df_anonymized = df_anonymized[new_columns]
```
