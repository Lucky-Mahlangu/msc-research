```python
import networkx as nx
import matplotlib.pyplot as plt


# Specify the path to your data file
file_path = '../data/ground_truth/follower_data/follower_data_links_anonymized_250.txt'

# Read data from the file
with open(file_path, 'r') as file:
    edges = [line.strip().split() for line in file]

# Create a directed graph
G = nx.DiGraph()

# Add reversed edges to the graph
reversed_edges = [(edge[1], edge[0]) for edge in edges]
G.add_edges_from(reversed_edges)


# Increase the plot size
plt.figure(figsize=(20, 15)) 


# Generate the layout with spring_layout
pos = nx.spring_layout(G)

# Node color
node_colors = ['#1f78b4' if node in G.pred[node] else '#33a02c' for node in G.nodes]

nx.draw(G, pos, with_labels=True, font_size=12, node_size=50, font_color='black', font_weight='bold', edge_color='gray', alpha=0.7, arrowsize=10, node_color=node_colors)

# Display the plot
plt.title('Visualisation of the generated social network sample data', fontsize=40)
#plt.savefig('social_network_plot.png') 
plt.show()
```
