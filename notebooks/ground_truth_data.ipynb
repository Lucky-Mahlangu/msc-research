```python
beauty_and_makeup = pd.read_excel("../data/ground_truth/xlsx_data/top_beauty_and_makeup_influencers.xlsx")
entertainment = pd.read_excel("../data/ground_truth/xlsx_data/top_entertainment_influencers.xlsx")
family = pd.read_excel("../data/ground_truth/xlsx_data/top_family_influencers.xlsx")
fashion = pd.read_excel("../data/ground_truth/xlsx_data/top_fashion_influencers.xlsx")
food = pd.read_excel("../data/ground_truth/xlsx_data/top_food_influencers.xlsx")
gaming = pd.read_excel("../data/ground_truth/xlsx_data/top_gaming_influencers.xlsx")
health_sports_fitness = pd.read_excel("../data/ground_truth/xlsx_data/top_health_sports_fitness_influencers.xlsx")
lifestyle = pd.read_excel("../data/ground_truth/xlsx_data/top_lifestyle_influencers.xlsx")
photography = pd.read_excel("../data/ground_truth/xlsx_data/top_photography_influencers.xlsx")
technology = pd.read_excel("../data/ground_truth/xlsx_data/top_technology_influencers.xlsx")
travel = pd.read_excel("../data/ground_truth/xlsx_data/top_travel_influencers.xlsx")
```

```python
category_list = ["beauty_and_makeup", "entertainment", "family", "fashion", \
              "food", "gaming", "health_sports_fitness", "lifestyle", \
              "photography", "technology", "travel"
              ]
```

```python
df = [beauty_and_makeup, entertainment, family, fashion, food,\
                gaming, health_sports_fitness, lifestyle,\
                photography, technology, travel]
```

```python
def extract_instagram_handle(df: pd.DataFrame, category: str) -> pd.DataFrame:
    """
    Extract Instagram handles.

    Parameters:
    - df (pd.DataFrame): The input DataFrame containing a 'profile' column.
    - category (str): The category string to assign to the DataFrame.

    Returns:
    - pd.DataFrame: A new DataFrame.
    """
    df_clean = df.copy()
    df_clean['instagramHandle'] = df_clean['profile'].str.extract(r'viralpitch.co/instagram/([^/]+)/')
    df_clean['instagramProfile'] = "https://www.instagram.com/" + df_clean['instagramHandle'] + "/"
    df_clean = df_clean[~df_clean['instagramHandle'].isna()]
    df_clean = df_clean.reset_index(drop=True)
    df_clean['topK'] = df_clean.index + 1
    df_clean = df_clean.iloc[:-1]
    df_clean['domain'] = category
    df_clean['topInfluencer'] = 1
    return df_clean
```

```python
result_df = pd.DataFrame({})
for i in range(len(category_list)):
    result_df = pd.concat([result_df, extract_instagram_handle(df[i], category_list[i])])
```

```python
# Specify the folder path where your Excel files are located
folder_path = '../data/ground_truth/follower_data/'

# Get a list of all files in the folder
all_files = os.listdir(folder_path)

# Filter the list to include only Excel files (assuming they have '.xlsx' extension)
excel_files = [file for file in all_files if file.endswith('.xlsx')]

# Create an empty dictionary to store DataFrames
dfs = {}

# Loop through each Excel file and read it into a DataFrame
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    # Assuming the first sheet in each Excel file is the one you want to read
    dfs[file] = pd.read_excel(file_path, sheet_name=0)
```
